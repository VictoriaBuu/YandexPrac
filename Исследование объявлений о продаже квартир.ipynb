### Предобработка данных


```python
data.isna().sum()
data['parks_around3000'] = data['parks_around3000'].fillna('0')
data['ponds_around3000'] = data['ponds_around3000'].fillna('0')
data['balcony'] = data['balcony'].fillna('0')
#data.loc[(data['parks_around3000'] >=1) & data['parks_nearest'].isna()]
```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Заполним нулями

Такие значение, как парки, пруды в районе 3км заполнили 0, так как можно предположить, что по аналогии с балконом, люди пропустили заполнение этих пунктов в связи с их отстуствием, по-скольку 3 км не такой и большой радиус, то вполне вероятно, что парки и пруды могут не встречаться в областях зоны где размещены объявления.


```python
for t in data['rooms'].unique():
    data.loc[(data['rooms'] == t) & (data['living_area'].isna()), 'living_area'] = \
    data.loc[(data['rooms'] == t), 'living_area'].mean()

for t in data['rooms'].unique():
    data.loc[(data['rooms'] == t) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
    data.loc[(data['rooms'] == t), 'kitchen_area'].mean()


```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Согласен, тут для заполнения можно опереться на комнаты

Пропуски заполняем средним значением, пропущенные значения столбцов 'living_area' и 'kitchen_area' - будем заполнять используя столбец 'rooms'. Т.е. пропуски будут заполнены средними значениями квартир с таким же количеством комнат. 

```python

data = data[data['locality_name'].notna()]
data= data[data['floors_total'].notna()]
data.isna().sum() 



```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Отбросим

По-скольку угадать название, где расположено объявление не возможно, то логично будет удалить эти данные.
Также удалим данные из столбца 'floors_total', так как их немного и они нам не пригодятся.


```python

data['balcony'] = data['balcony'].astype(int)
data['ponds_around3000'] = data['ponds_around3000'].astype(int)
data['parks_around3000'] = data['parks_around3000'].astype(int)
data['last_price'] = data['last_price'].astype(int)
data['first_day_exposition'] = pd.to_datetime(
    data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S'
)
data['is_apartment'] = data['is_apartment'].astype(bool)
data['floors_total']= data['floors_total'].astype(int)
data.info()

print(data.head(10))
```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Такие типы данных подойдут им лучше

Изменение типов данных помогает проводить необходимые операции с числами,  к примеру, если в столбце 'floors_total' будет тип данных objects, это будет значить, что там есть нечисловые значения, которые потом помешают проводить расчеты.
 Это касается и даты, если мы оставим исходное значение дат, а не изменим  его на привычные datetime64[ns], то с ним нельзя будет проводить арифметические операции.

```python

data['locality_name'] = data['locality_name'].replace(to_replace = 'посёлок', value = '', regex = True)
data['locality_name'] = data['locality_name'].replace(to_replace = 'деревня', value = '', regex = True)
data['locality_name'] = data['locality_name'].replace(to_replace = 'садовое товарищество', value = '', regex = True)
data['locality_name'] = data['locality_name'].replace(to_replace = 'село', value = '', regex = True)

data['locality_name'].nunique()



```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Молодец, что обработала неявные дубликаты

Теперь мы удаляем все названия содержащие обозначение типа нас.пункта., потому что нам эти данные не понадобятся, поскольку это не повлияет на конечный результат и зависимость стоимости квартиры.

### Посчитайте и добавьте в таблицу новые столбцы

```python
data['price_for_one_m'] = data['last_price']/ data['total_area']

print(data)







```

```python
data['weekday'] = data['first_day_exposition'].dt.weekday
def day_expos(row):
    if row['weekday'] == 0:
        return 'понедельник'
    elif row['weekday'] == 1:
        return 'вторник'
    elif row['weekday'] == 2:
        return 'среда'
    elif row['weekday'] == 3:
        return 'четверг'
    elif row['weekday'] == 4:
        return 'пятница'
    elif row['weekday'] == 5:
        return 'суббота'
    elif row['weekday'] == 6:
        return 'воскресение'
data['day_expos'] = data.apply(day_expos, axis = 1)

import calendar
data['month'] =data['first_day_exposition'].dt.month
data['month'] = data['month'].apply(lambda x:calendar.month_abbr[x])

data['year'] = data['first_day_exposition'].dt.year
#print(data)
```

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Рабочая функция

```python
def floor_num(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor']  == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['floor_num'] = data.apply(floor_num, axis = 1)
data[['floor', 'floor_num', 'floors_total']].head(5)
#print(data)
```
